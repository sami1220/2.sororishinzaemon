<span style="font-size: 25px">【問題1】100日目の米粒の数</span>

```python
# 米粒の数
rice = 1
# トータルで米を貰う日数
total_day = 100
# トータルで米をもらった数
total_rice = 0

# グラフの初期値
graph = []

for i in range(1, int(total_day + 1)):
    total_rice += rice
    rice *= 2
    graph.append(total_rice)

# ①100日目には合計で米粒をいくつもらっているのかを計算するコード
print(total_rice)

# ②グラフ （縦軸が米の数、横軸が日数）
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("sororishinzaemon get rice")
plt.xlabel("Days")
plt.ylabel("Grain of rice")
plt.plot(graph)
plt.show()
```

<span style="font-size: 25px">【問題3】貰える米で何人が何日生活できるか</span>

この米があれば何人が何日生活できるか計算する関数を作成してください。コースを受講する期間に貰える米粒で、このクラスにいる人数であれば、何日間生活できるかを出力してみましょう。


関数の最低限の入力は 米粒の数(int) と 人数(int) 、出力は 生活できる日数(int) です。関数名やコメントなどは分かりやすく考えて作成してください。


ここでは1日に必要な米粒の数は提示しません。自分なりに調べて、計算式に落とし込んでください。どう調べ、何を根拠にしたかもコメントを残しておいてください。

```python
import math

def life_days(rice_grain, num):
    """もらった米の数と合計の人数で、何日間生活できるか計算する関数
    Parameteres
    -----------
    rice_grain
        もらう米粒の合計
    num
        人数
    person　= 5840 int
        1日の必要となる米粒の数(5840 = 146,000mg[一合弱])
    Returns
    -------
    answer = rice_grain / person
        ①もらった米数で何日間生活できるのか（１人）
    answer2 = answer / num
        ②全員でどれだけの日数生活できるのか
    """

    person = 5840
    
    answer = rice_grain / person
    answer2 = answer / num
    
    return math.floor(answer2)

print("{}日生きる事ができます".format(life_days(1267650600228229401496703205375, 15)))
```
